```python
#import module
import xmltodict
import requests
import rauth
import pandas as pd
import numpy as np
from goodreads import client
```

```python
#melminta izin akses
dts = client.GoodreadsClient('HQgj4RR32KTrSkaq7NEg', '3FdKHONgsCuxh69r2Lw0YhX1PzTbeKt3RGGSxKrCU')
dts.authenticate()
```

```python
#mendefenisikan list/kolom
Id_Buku=[]
Isbn=[]
Judul_Buku=[]
Penulis=[]
Tahun_Publikasi=[]
Total_Ratings=[]
Rataan_Rating=[]
Jumlah_Halaman=[]
Jumlah_Review=[]
Kode_Bahasa=[]
Format_Buku=[]
Ebook=[]
Penerbit=[]
```

```python
#crawling data
for i in range (1,10000):
    try:
        book = dts.book(i)
  
        Id = int(book.gid)
        ratingss_count = int(book.ratings_count)
        av_rating = float(book.average_rating)
        jml_hlm = int(book.num_pages)
        jml_review = int(book.text_reviews_count)
    
        Id_Buku +=[Id]
        Isbn +=[book.isbn]
        Judul_Buku += [book.title]
        Penulis +=[book.authors]
        Tahun_Publikasi +=[book.publication_date[2]]
        Total_Ratings += [ratingss_count]
        Rataan_Rating +=[av_rating]
        Jumlah_Halaman +=[jml_hlm]
        Jumlah_Review +=[jml_review]
        Kode_Bahasa +=[book.language_code]
        Format_Buku +=[book.format]
        Ebook +=[book.is_ebook]
        Penerbit +=[book.publisher]
    except Exception:
        pass
```

```python
Daftar_Buku = pd.DataFrame({'Id_Buku': Id_Buku,
                             'Isbn': Isbn,
                             'Judul_Buku': Judul_Buku,
                             'Penulis': Penulis,
                             'Tahun_Publikasi': Tahun_Publikasi,
                             'Total_Ratings': Total_Ratings,
                             'Rataan_Rating': Rataan_Rating,
                             'Jumlah_Halaman': Jumlah_Halaman,
                             'Jumlah_Review': Jumlah_Review,
                             'Kode_Bahasa': Kode_Bahasa,
                             'Format_Buku': Format_Buku,
                             'Ebook': Ebook,
                             'Penerbit': Penerbit})
```

```python
Daftar_Buku = Daftar_Buku[['Id_Buku',
                             'Isbn',
                             'Judul_Buku',
                             'Penulis',
                             'Tahun_Publikasi',
                             'Total_Ratings',
                             'Rataan_Rating',
                             'Jumlah_Halaman',
                             'Jumlah_Review',
                             'Kode_Bahasa',
                             'Format_Buku',
                             'Ebook',
                             'Penerbit']]
```
