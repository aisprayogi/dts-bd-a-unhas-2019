```python
import pandas as pd
```

```python
df1 = pd.read_csv('movie_ratings10-14.csv')
df2 = pd.read_csv('movie_ratings15-19.csv')
```

```python
#join multiple csv
result = df1.append(df2, ignore_index=True, sort=False)
```

```python
# new data frame with split value columns 
new = result["genres"].str.split(",", n = 2, expand = True) 
new
```

```python
# making separate first name column from new data frame 
result["First genre"]= new[0] 
# making separate second name column from new data frame 
result["Second genre"]= new[1] 
# making separate last name column from new data frame 
result["Last genre"]= new[2] 
```

```python
# Dropping old Name columns 
result.drop(columns =["genres"], inplace = True) 
```

```python
result.loc[:, 'First genre'] = result['First genre'].str[1:]
```
