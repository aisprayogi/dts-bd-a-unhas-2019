```python
import pandas as pd
```

```python
df1 = pd.read_csv('movie_ratings10-14.csv')
df2 = pd.read_csv('movie_ratings15-19.csv')
```

```python
#join multiple csv
result = df1.append(df2, ignore_index=True, sort=False)
result.head()
```

```python
# new data frame with split value columns 
new = result["genres"].str.split(",", n = 2, expand = True) 
new
```

```python
# making separate first name column from new data frame 
result["First genre"]= new[0] 
# making separate second name column from new data frame 
result["Second genre"]= new[1] 
# making separate last name column from new data frame 
result["Last genre"]= new[2] 
```

```python
# Dropping old Name columns 
result.drop(columns =["genres"], inplace = True) 
```

```python
result.loc[:, 'First genre'] = result['First genre'].str[1:]
```

```python
result
```

```python
del result['Unnamed: 0']
```

```python
result
```

```python
# new data frame with split value columns directors
new2 = result["directors"].str.split(",", n = 2, expand = True) 
new2
```

```python
# making separate first name column from new data frame 
result["First Director"]= new2[0] 
# making separate second name column from new data frame 
result["Second Director"]= new2[1] 
# making separate last name column from new data frame 
result["Last Director"]= new2[2] 
```

```python
# Dropping old Name columns 
result.drop(columns =["directors"], inplace = True) 
```

```python
result
```

```python
del result['Last Director']
```

```python
del result['Second Director']
```

```python
result['Director']=new2[0]
```

```python
# Dropping old Name columns 
result.drop(columns =["First Director"], inplace = True) 
```

```python
for i in range(len(result['Director'])):
    if result['Director'][i][0]=='\n':
        result['Director'][i]=result['Director'][i][1:]
```

```python
#bersihkan whitespace jika ada pada kolom value untuk genre
result['Last genre'] = result['Last genre'].str.strip()
result['First genre'] = result['First genre'].str.strip()
```

```python
result.drop_duplicates(subset =['movie', 'year'], 
                     keep = 'first', inplace = True)
```

```python
result.to_csv('data_fix.csv')
```
